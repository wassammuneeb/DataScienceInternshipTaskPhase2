```python
#Import Libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the Dataset

df = sns.load_dataset('iris')

```

```python
# Check Dataset Structure
print("Shape of the dataset:", df.shape)
print("Column names:", df.columns.tolist())
print("First 5 rows:")
print(df.head())
```

```python
# Check for Missing Data
print("Missing values in each column:")
print(df.isnull().sum())

```

```python
# Summary Statistics
print("Summary statistics:")
print(df.describe())

```

```python
# Scatter Plot
sns.pairplot(df, hue="species", palette="Set2")
plt.suptitle("Scatter Plot Matrix of Iris Dataset", y=1.02)
plt.show()
```

```python
# Histograms
df.hist(figsize=(10, 8), bins=15, edgecolor='black')
plt.suptitle("Histograms of Each Feature", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
# Box Plots
for col in df.columns[:-1]:  # skip 'species'
    sns.boxplot(x='species', y=col, data=df)
    plt.title(f"Boxplot of {col} by Species")
    plt.show()
```
